#### Apredizado Supervisionado 

```python
# Simple Linear Regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

```python
print(y)
print(x)
```

```python
model = LinearRegression(fit_intercept=True)
model
```

```python
# reformulacao do array x
X = x[:, np.newaxis]
X
```

```python
# fit model to data
model.fit(X, y)
```

```python
model.coef_
```

```python
# Predict labels for unknown data
# "new data" will be a grid of x values, and we will ask what y values the model predicts
xfit = np.linspace(-1, 11)
xfit
```

```python
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit)
```
